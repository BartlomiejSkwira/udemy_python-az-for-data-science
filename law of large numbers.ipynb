```python
from numpy.random import randn

n = 1
counter = 0
mean = 0
print("starting")

for rand in randn(100000):
    if -1 < rand < 1:
        counter = counter + 1
    mean = counter / n
    if n % 1000 == 0:
        print("i ==", rand, "n ==", n, "mean ==", mean)
    n = n + 1
    
print("ending")
```
