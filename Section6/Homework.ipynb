```python
import pandas as pd
pd.set_option('max_rows', 7)

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 8
import seaborn as sns
```

```python
movies = pd.read_csv("../data/Section6-Homework-Data.csv", encoding = "ISO-8859-1")
movies.head()
```

```python
movies.describe()
```

```python
movies.info()
```

```python
selected_genres = ['action', 'comedy', 'adventure', 'animation', 
                   'drama']
selected_studios = ['Buena Vista Studios', 'Sony', 'Universal',
                    'WB', 'Paramount Pictures', 'Fox']
filter = movies.Genre.isin(selected_genres) & movies.Studio.isin(selected_studios)
sns.set(style='darkgrid', palette='muted', color_codes=True)
w = sns.boxplot(data=movies[filter], 
                x='Genre', y='Gross % US', 
                color='lightgrey',
                orient='v',)
                #showFilter=False)
plt.setp(w.artists, alpha=0.5)
w = sns.swarmplot(data=movies[filter], 
                x='Genre', y='Gross % US', hue='Studio')
w.set_xlabel('Genre', fontsize=20)
w.set_ylabel('Gross % US', fontsize=20)
w.set_title('Domestic Gross % by Genre', fontsize=40)
w.legend(bbox_to_anchor=(1, 1), loc=2)
```
